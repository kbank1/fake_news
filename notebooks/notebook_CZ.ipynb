```python
def fix_titles_longer_than_texts(datasets):
    for data in datasets:
        # Identify rows where the title is longer than the text
        condition = data['title_length'] > data['text_length']
        
        # Swap the title and text columns where the condition is True
        data.loc[condition, ['title', 'text']] = data.loc[condition, ['text', 'title']].values
        
        # Recalculate the title and text lengths
        data['title_length'] = data['title'].apply(len)
        data['text_length'] = data['text'].apply(len)
        
    return datasets

# Apply the function to your datasets
datasets = fix_titles_longer_than_texts(datasets)

# Check the number of problematic rows after the fix
count_titles_longer_than_texts(datasets, dataset_names)
```
